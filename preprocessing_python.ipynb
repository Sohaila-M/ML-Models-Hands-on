```python
import pandas as pd
import scipy
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1= pd.read_excel('data for preprocessing.xlsx')
```

```python
df1.isnull().sum()
#Replacing - With NULL
df1.replace('-',pd.NA)
```

```python
#Converting Sent time to time only and Calculating the mode value
df1['Sent time'] = pd.to_datetime(df1['Sent time'].astype(str), errors='coerce')
df1['Sent time'] = df1['Sent time'].dt.time #extracting time only
mode_sent_time = df1['Sent time'].dropna().mode()
mode_sent_time_value = mode_sent_time[0]
print(mode_sent_time)
print(mode_sent_time_value)
```

```python
#Replacing Sent Time NULL Values with mode
df1['Sent time'] = df1['Sent time'].fillna(mode_sent_time_value)
#Replacing the value 'intelligent delivery' with the mode
df1.replace(to_replace='intelligent delivery', value=mode_sent_time_value)
```

```python
print(df1['Sent time'].isnull().sum())
print(df1.loc[114, 'Sent time'])
```

```python
#Dropping End date Column
df2=df1.drop('End date', axis=1)
```

```python
df2
```

```python
# Recipients Column is connected to Notification opens column so we must calculate the ratio between them

df2['Notification opens'] = pd.to_numeric(df2['Notification opens'], errors='coerce')
df2['Recipients'] = pd.to_numeric(df2['Recipients'], errors='coerce')

mean_open_ratio = df2['Notification opens'].dropna() / df2['Recipients']
mean_open_ratio = mean_open_ratio.mean()  # Calculate the mean ratio
print(mean_open_ratio)
```

```python
#filling the missing values based on the relationship between Notification opens and Recipients

df3=df2.copy()

df3.loc[df3['Notification opens'].isna(), 'Notification opens'] = df3['Recipients'] * mean_open_ratio

print(df3['Notification opens'].isnull().sum())
```

```python

#checking if there's a relationship between Notification opens and App opens
df3['Notification opens'] = pd.to_numeric(df3['Notification opens'], errors='coerce')
df3['App opens'] = pd.to_numeric(df3['App opens'], errors='coerce')

correlation = df3['Notification opens'].corr(df3['App opens'])

print(correlation)
```

```python
#Replacing NULL values of App opens with mean

df4=df3.copy()
mean_app_opens = df4['App opens'].dropna().mean()
df4['App opens'] = df1['App opens'].fillna(mean_app_opens)

print(df4['App opens'].isnull().sum())
```

```python
#Many Rows in Clicks Contain - so we'll replace it with NULL

df5=df4.copy()

df5['Clicks'] = df5['Clicks'].replace('-', pd.NA)

print(df5['Clicks'].isnull().sum())

#dropping Clicks

df6=df5.drop('Clicks', axis=1)

print(df6.isnull().sum())
```

```python
df7=df6.drop('Day1 Conv.', axis=1)
print(df7.isnull().sum())
```

```python
df8=df7.copy()

df8['Control Group Volume'] = df8['Control Group Volume'].replace('-', pd.NA)

print(df8['Control Group Volume'].isnull().sum())
```

```python
df9=df8.copy()

df9['Control Group Conv'] = df9['Control Group Conv'].replace('-', pd.NA)

print(df9['Control Group Conv'].isnull().sum())
```

```python
df10=df9.copy()

df10['Control Group Rev'] = df10['Control Group Rev'].replace('-', pd.NA)

print(df10['Control Group Rev'].isnull().sum())
```

```python
#Forward Filling Voucher Name

df11=df10.copy()

df11['Voucher Name']=df11['Voucher Name'].fillna(method='ffill')

print(df11['Voucher Name'].isnull().sum())
```

```python
df12=df11.copy()

df12['Registrations'] = df12['Registrations'].replace('-', pd.NA)

print(df12['Registrations'].isnull().sum()) # o/p:1727
#Dropping Registrations column
df12=df12.drop('Registrations', axis=1)

print(df12.isnull().sum())
```

```python
df13=df12.copy()
df13['Campaing Revenue'] = pd.to_numeric(df13['Campaing Revenue'], errors='coerce')
mean_CampaingRevenue = df13['Campaing Revenue'].dropna().mean()
df13['Campaing Revenue'] = df13['Campaing Revenue'].fillna(mean_CampaingRevenue)

print(df13['Campaing Revenue'].isnull().sum())

df13['Conversions'] = pd.to_numeric(df13['Conversions'], errors='coerce')
mean_CampaingRevenue = df13['Conversions'].dropna().mean()
df13['Conversions'] = df13['Conversions'].fillna(mean_CampaingRevenue)

print(df13['Conversions'].isnull().sum())
```

```python
df14=df13.copy()
#Replacing NULL values of Control Group Volume with mean
mean_ControlGroupVolume = df14['Control Group Volume'].dropna().mean()
df14['Control Group Volume'] = df14['Control Group Volume'].fillna(mean_app_opens)
print(df14['Control Group Volume'].isnull().sum())
#Replacing NULL values of Control Group Conv with mean
mean_ControlGroupConv = df14['Control Group Conv'].dropna().mean()
df14['Control Group Conv'] = df14['Control Group Conv'].fillna(mean_ControlGroupConv)
print(df14['Control Group Conv'].isnull().sum())
```

```python
#  remove 'SAR' from strings and clean the column
df14['Control Group Rev'] = (
    df14['Control Group Rev']
    .astype(str)                        # ensure it's string
    .str.replace('SAR', '', regex=False)
    .str.replace(',', '')              # optional: remove commas
    .str.strip()
)

# Convert to numeric
df14['Control Group Rev'] = pd.to_numeric(df14['Control Group Rev'], errors='coerce')

# Fill NULL with the mean
mean_ControlGroupRev = df14['Control Group Rev'].mean()
print("Mean:", mean_ControlGroupRev)
df14['Control Group Rev'] = df14['Control Group Rev'].fillna(mean_ControlGroupRev)
print(df14['Control Group Rev'].tail())

```
